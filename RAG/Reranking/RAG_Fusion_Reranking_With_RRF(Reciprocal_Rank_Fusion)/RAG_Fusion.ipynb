```python
import textwrap
def wrap_text(text, width=90): #preserve_newlines
    # Split the input text into lines based on newline characters
    lines = text.split('\n') # wrap_text function splits the input text by newlines (\n)

    # Wrap each line individually
    wrapped_lines = [textwrap.fill(line, width=width) for line in lines] # wraps each line individually to a width of 90 characters

    # Join the wrapped lines back together using newline characters
    wrapped_text = '\n'.join(wrapped_lines) # joined back with newlines (\n) to maintain the paragraph structure

    return wrapped_text
```

## With RagFusion

```python
from langchain.schema.output_parser import StrOutputParser
from langchain.prompts import SystemMessagePromptTemplate, HumanMessagePromptTemplate
from langchain.prompts import ChatMessagePromptTemplate, PromptTemplate
```

```python
prompt = ChatPromptTemplate(
    input_variables=['original_query'],  # The variable to pass user input
    messages=[
        # System message to define AI behavior
        SystemMessagePromptTemplate(
            prompt=PromptTemplate(
                input_variables=[],  # No variables for system-level instructions
                template=(
                    "You are a helpful assistant that generates multiple search queries "
                    "based on a single input query."
                )
            )
        ),
        # Human message representing the user's input
        HumanMessagePromptTemplate(
            prompt=PromptTemplate(
                input_variables=['original_query'],  # Variable passed by the user
                template=(
                    "Generate multiple search queries related to: {original_query}\n"
                    "OUTPUT (4 queries):"
                )
            )
        )
    ]
)
```

```python
original_query = "universal studios Singapore"
```

```python
generate_queries = (
    prompt | llm | StrOutputParser() | (lambda x: x.split("\n"))
)
```

```python
generate_queries
```

```python
from langchain.load import dumps, loads


def reciprocal_rank_fusion(results: list[list], k=60):
    fused_scores = {} # key(doc), value(score) store
    for docs in results:
        # Assumes the docs are returned in sorted order of relevance
        for rank, doc in enumerate(docs): # calculating score for each and every doc using rank and k=60 and store in fused_scores
            doc_str = dumps(doc) # serializes doc into a string
            if doc_str not in fused_scores:
                fused_scores[doc_str] = 0
            previous_score = fused_scores[doc_str]
            fused_scores[doc_str] += 1 / (rank + k)

    reranked_results = [ # reranked calculated fused_scores in descending order
        (loads(doc), score) for doc, score in sorted(fused_scores.items(), key=lambda x: x[1], reverse=True) # reverse=True ensures that the sorting is in descending order
    ]    # loads(doc): Deserialize the document string (doc_str) back into its original format
    return reranked_results
```

```python
ragfusion_chain = generate_queries | retriever.map() | reciprocal_rank_fusion
```

```python
langchain.debug = True
```

```python
ragfusion_chain.input_schema.schema()
```

```python
ragfusion_chain.invoke({"question": original_query})
```

```python
from langchain.schema.runnable import RunnablePassthrough
template = """Answer the question based only on the following context:
{context}

Question: {question}
"""
prompt = ChatPromptTemplate.from_template(template)

full_rag_fusion_chain = (
    {
        "context": ragfusion_chain,
        "question": RunnablePassthrough()
    }
    | prompt
    | llm
    | StrOutputParser()
)
```
